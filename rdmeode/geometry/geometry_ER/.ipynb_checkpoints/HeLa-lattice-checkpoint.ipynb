# Build RDME lattice geometry from electron density


First we build a simple helper function to allow us to interactively choose the minimum density to associate a site type with a particular compartment.

```python
binaryLattices = dict()

def makeThreshold2(reg, c=None, th=None, z=None):
    density = densityData[reg]
    v0=np.min(density)
    v1=np.max(density)
    binaryLattices[reg] = np.zeros(density.shape, dtype=bool)
    bim = binaryLattices[reg]
    def fn(threshold, z_index, closing):
        fig, ax = plt.subplots(figsize=(9,9))
        bim[...] = B.closing(density > threshold, se=B.sphereStructElem(closing))
        nx,ny = bim.shape[0], bim.shape[2]
        ax.imshow(density[:,z_index,:], vmin=v0, vmax=v1, cmap='Blues', aspect="auto", interpolation="nearest")
        if np.any(bim[:,z_index,:]):
            ax.add_patch(boundaryPath(bim[:,z_index,:]))
        ax.set(xlim=(0,nx),ylim=(0,ny), title="{} threshold={:.5e}".format(reg.name,threshold))
        return None
    ipw.interact(fn, 
                 closing=ipw.IntSlider(value=c, min=1, max=10,continuous_update=False),
                 threshold=ipw.FloatSlider(value=th, min=0.0,max=1,step=0.0001, continuous_update=False),
                 z_index=ipw.IntSlider(value=z, min=0,max=density.shape[1]-1, continuous_update=False))
    return None

def makeThreshold(reg, d=None, e=None, th=None, z=None):
    density = densityData[reg]
    v0=np.min(density)
    v1=np.max(density)
    binaryLattices[reg] = np.zeros(density.shape, dtype=bool)
    bim = binaryLattices[reg]
    def fn(threshold, z_index, dilation, erosion):
        fig, ax = plt.subplots(figsize=(9,9))
        bim[...] = B.erode(B.dilate(density > threshold, dilation),erosion)
        nx,ny = bim.shape[0], bim.shape[2]
        ax.imshow(density[:,z_index,:], vmin=v0, vmax=v1, cmap='Blues', aspect="auto", interpolation="nearest")
        if np.any(bim[:,z_index,:]):
            ax.add_patch(boundaryPath(bim[:,z_index,:]))
        ax.set(xlim=(0,nx),ylim=(0,ny), title="{} threshold={:.5e}".format(reg.name,threshold))
        return None
    ipw.interact(fn, 
                 dilation=ipw.IntSlider(value=d, min=1, max=10,continuous_update=False),
                 erosion=ipw.IntSlider(value=e, min=1, max=10,continuous_update=False),
                 threshold=ipw.FloatSlider(value=th, min=0.0,max=1,step=0.0001, continuous_update=False),
                 z_index=ipw.IntSlider(value=z, min=0,max=density.shape[1]-1, continuous_update=False))
    return None
```

# Construct simulation volume

Now we must build the compartments corrisponding to the cytoplasm and nucleoplasm. Since we do not have cryo-ET density maps describing these regions, we use the density we have to choose a bounding volume.

```python
microtubialReg = binaryLattices[microtubules]
actinReg = binaryLattices[actin]
erReg = binaryLattices[er]
ssuReg = binaryLattices[ssu]
lsuReg = binaryLattices[lsu]
npcReg = binaryLattices[npc]
```

```python
dist, lsus, ssus, topDists, ssuPos = dict(), set(), set(), [], []

ssuDensity = densityData[ssu]*ssuReg
lsuDensity = densityData[lsu]*lsuReg

md = max(1,int(13e-9/sim.latticeSpacing))
print("min ribosome peak distance: ", md)
ssuLoc = skfe.peak_local_max(ssuDensity, threshold_rel=1e-9, min_distance=md)
lsuLoc = skfe.peak_local_max(lsuDensity, threshold_rel=1e-9, min_distance=md)

for (i,v0), (j,v1) in itertools.product(enumerate(ssuLoc), enumerate(lsuLoc)):
    dist[(i,j)] = np.sqrt(np.sum((v0-v1)**2))

lsus, ssus, ssuPos, lsuPos = set(), set(), [], []

ribosomes = 222
md = max(1,int(16e-9/sim.latticeSpacing))

pairs = sorted([ (s,l,d) for (s,l),d in dist.items() if d < md ], key=lambda x: (x[2]-12.7e-9/sim.latticeSpacing)**2)

for s,l,_ in pairs:
    if len(ssuPos) == ribosomes:
        break
    if s not in ssus and l not in lsus:
        lsus.add(l)
        ssus.add(s)
        ssuPos.append(tuple(map(int,ssuLoc[s,:])))
        lsuPos.append(tuple(map(int,lsuLoc[l,:])))
ssuPosold = ssuPos
```

Now from this convex hull of the em density, we will partition it into cytoplasm, nucleoplasm, and nuclear membrane. We use the positions of the nuclear pores to find a reasonable location and orientation of a plane which will represent the membrane.

```python
data = np.argwhere(npcReg)
pt = np.mean(data,axis=0)
r = 5500e-9/sim.latticeSpacing
dr = 40e-9/sim.latticeSpacing
ptdir = pt/np.sqrt(pt@pt)
pt -= 4*ptdir - 24e-9/sim.latticeSpacing

phi = np.pi/4
theta = np.pi/8
```

```python
ctr = pt + np.array([r*np.sin(phi)*np.sin(theta), r*np.cos(theta), r*np.cos(phi)*np.sin(theta)])
nucleoplasmReg = B.ellipsoid(B.xs, radius=r, center=ctr)
nuclearenvelopeReg = B.ellipsoid(B.xs, radius=r+dr, center=ctr) & ~nucleoplasmReg
nuclearenvelopesolidReg = nuclearenvelopeReg.copy()
lbl, _ = ndi.label(nuclearenvelopeReg & ~npcReg)
cts = np.bincount(lbl.ravel())
sidx = np.argsort(cts)[::-1]
nuclearenvelopeReg = lbl == sidx[1]
poreReg = (lbl == sidx[2]) | (lbl == sidx[3]) | (lbl == sidx[4]) | (lbl == sidx[5])
```

```python
everything = microtubialReg|actinReg|erReg|ssuReg|lsuReg|npcReg
cHull = B.convexHull(everything)
```

```python
#B.showBinaryLattices(everything=everything, hull=cHull)
```

```python
nHull = B.dilate((ndi.gaussian_filter(densityData[npc], 10) > 0.0009) & nucleoplasmReg, 2)
hull = B.closing(nHull|cHull, se=B.sphereStructElem(10))
```

```python
#boundaryReg = hull & ~B.erode(hull, 1)
```

```python
nucleoplasmReg &= hull
```

```python
nuclearenvelopeReg &= hull
nuclearenvelopesolidReg &= hull
```

```python
simpleLattice=np.zeros_like(sim.siteLattice)

obstruction = npcReg | nuclearenvelopeReg | microtubialReg | actinReg | erReg | lsuReg

simpleLattices = [hull,
                  nucleoplasmReg, 
                  poreReg,
                  obstruction,
                  ssuReg, ]
                  #boundaryReg]

simpleLatticeNames = ["extracellular", "cytoplasm", "nucleoplasm",  "poreChannel", "obstruction", "ssu", ]#"boundary"]
for i,reg in enumerate(simpleLattices):
    simpleLattice[reg] = i+1
```

```python
sim.siteLattice[...]=0
B.compose(
    (cytoplasm, hull),
    (nucleoplasm, nucleoplasmReg),
    (microtubules, microtubialReg), 
    (actin, actinReg), 
    (er, erReg), 
    (lsu, lsuReg),
    (ssu, ssuReg), 
    (nuclearEnv, nuclearenvelopeReg),
    (npc,npcReg), 
    #(boundary, boundaryReg),
    (pore, poreReg)
)
```

```python
xtr = np.diff(1.0*(np.sum(sim.siteLattice, axis=(1,2))>0))
ytr = np.diff(1.0*(np.sum(sim.siteLattice, axis=(0,2))>0))
ztr = np.diff(1.0*(np.sum(sim.siteLattice, axis=(0,1))>0))
```

```python
plt.plot(xtr,label='x')
plt.plot(ytr,label='y')
plt.plot(ztr,label='z')
plt.legend()
```

```python
def dimfn(u):
    v = (u+1)%3
    w = (u+2)%3
    ct = np.sum(sim.siteLattice, axis=(v,w))
    u0 = np.where(ct>0)[0][0] 
    uw = np.sum(ct>0)
    uw32 = int(32*np.ceil(uw/32))
    spc = uw32-uw
    pad0 = spc//2
    return u0, uw, pad0, uw32

x0, xw, px0, nx = dimfn(0)
y0, yw, py0, ny = dimfn(1)
z0, zw, pz0, nz = dimfn(2)

nucEnvChopped = np.zeros_like(sim.siteLattice)
chopped = np.array(sim.siteLattice)
simpleChopped = np.array(simpleLattice)
sim.resizeLattice((nx,ny,nz),latticeSpacing)
sim.siteLattice[px0:px0+xw,py0:py0+yw,pz0:pz0+zw] = chopped[x0:x0+xw, y0:y0+yw, z0:z0+zw]
simpleLattice[...] = 0
simpleLattice[px0:px0+xw,py0:py0+yw,pz0:pz0+zw] = simpleChopped[x0:x0+xw, y0:y0+yw, z0:z0+zw]
nucEnvChopped[px0:px0+xw,py0:py0+yw,pz0:pz0+zw] = nuclearenvelopesolidReg[x0:x0+xw, y0:y0+yw, z0:z0+zw]
```

```python
vmdUnits = sim.latticeSpacing*1e10
resizeDx = px0-x0
resizeDy = py0-y0
resizeDz = pz0-z0
tclMatrix = ("set txMatrix {{1.0 0.0 0.0 dx} {0.0 0.0 1.0 dy} {0.0 1.0 0.0 dz} {0.0 0.0 0.0 1.0}}"
                    .replace("dx", str(int(vmdUnits*resizeDz)))
                    .replace("dy", str(int(vmdUnits*resizeDy)))
                    .replace("dz", str(int(vmdUnits*resizeDx))))
ssuPos = [(x+resizeDx, y+resizeDy, z+resizeDz) for x,y,z in ssuPos]
```

```python
pickle.dump(dict(labelLattice=np.array(sim.siteLattice, dtype=np.uint8), 
                 ssuPos=ssuPos,
                 lattice=simpleLattice,
                 closedNucEnv=nucEnvChopped,
                 names={x:i for i,x in enumerate(simpleLatticeNames)},
                 labelNames={x.name: i for i,x in enumerate(sim.regionList)}),
            open(simFile("siteLattice_{}e.p".format(int(sim.latticeSpacing*1e9))),"wb"))
   
with open(simFile("mesh_{}.tcl".format(int(sim.latticeSpacing*1e9))),"w") as tcl:
    tclTriangleMesh(sim, sim.siteLattice, file=tcl)
    print(tclMatrix, file=tcl)
```
