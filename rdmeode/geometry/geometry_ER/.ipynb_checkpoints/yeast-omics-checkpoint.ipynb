```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import qgrid
%matplotlib inline
from venv import *
```

Data taken from
 * [A quantitative estimation of the global translational activity in logarithmically growing yeast cells](https://bmcsystbiol.biomedcentral.com/articles/10.1186/1752-0509-2-87)

```python
abundanceData = pd.read_excel( dataFile("yeast-lit/1752-0509-2-87-S1.xls"),sheetname='Curated Dataset', index_col='ORF')
del abundanceData["Unnamed: 8"]
del abundanceData["Unnamed: 7"]
del abundanceData["Molecular Weight (Da)"]
del abundanceData["Length (Amino Acids)"]
abundanceData = abundanceData.rename(columns={"Protein Abundance (molecules per cell)": "Protein Abundance",
                              "Protein decay rate (min-1)": "Protein Decay Rate",
                              "mRNA Abundance (molecules per cell)": "mRNA Abundance"})
```

Data taken from
 * [Global Analysis of mRNA Isoform Half-Lives Reveals Stabilizing and Destabilizing Elements in Yeast](http://dx.doi.org/10.1016/j.cell.2013.12.026)

```python
mRNAdata = pd.read_excel( dataFile("yeast-lit/mmc1.xlsx"),sheetname='Sheet1', index_col='systematic name')
mRNAdata['mRNA Decay Rate'] = np.log(2)/mRNAdata["Half-Life           (in minutes)"]
del mRNAdata['chrom']
del mRNAdata['common name']
del mRNAdata['Simple Type']
del mRNAdata["Half-Life           (in minutes)"]
```

Gene ontology from
 * [Saccharomyces Genome Database](http://www.yeastgenome.org/download-data/curation)

```python
goData = pd.read_csv(dataFile("yeast-lit/go_slim_mapping.tab"), 
                     sep='\t', 
                     names=["ORF",
                            "Gene",
                            "SGDID",
                            "GO_Aspect",
                            "GO Slim term",
                            "GOID",
                            "Feature type"])

def goAspect(col, ch):
    terms = goData.loc[goData.GO_Aspect == ch, ['ORF','GO Slim term']].groupby("ORF")
    return terms.aggregate(lambda x: ' | '.join(x['GO Slim term'])).rename(columns={"GO Slim term": col})

goData = pd.concat([goAspect("Function", "F"), goAspect("Component", "C"), goAspect("Process", "P")], axis=1)
```

```python
yeastData = pd.concat([abundanceData, mRNAdata, goData], axis=1, join='inner')
yeastData['mRNA Decay Rate'] *= 1/60
yeastData['Protein Decay Rate'] *= 1/60
```

```python
yeastData['Transcription Rate'] = yeastData['mRNA Abundance']*yeastData['mRNA Decay Rate']
yeastData['Translation Rate'] = yeastData['Protein Abundance']*yeastData['Protein Decay Rate']/yeastData['mRNA Abundance']
```

```python
valid = ((yeastData['Protein Abundance'] > 0) & 
         (yeastData['mRNA Abundance'] > 0) & 
         (yeastData['Protein Decay Rate'] > 0) & 
         (yeastData['mRNA Decay Rate'] > 0))
```

```python
yeastData.describe()
```

```python
sns.distplot(np.log(2)/yeastData.loc[valid, 'mRNA Decay Rate'])
```

```python
qgrid.show_grid(yeastData.loc[valid,['Gene Name', 'Transcription Rate', 'mRNA Decay Rate', 'Translation Rate', 'Protein Decay Rate']])
```

```python
qgrid.show_grid(yeastData.loc[valid,['Gene Name', 'Description']])
```

```python
with sns.axes_style("white"):
    posprot = yeastData['Protein Abundance'] > 0
    posmrna = yeastData['mRNA Abundance'] > 0
    lgData = np.log10(yeastData.loc[posprot&posmrna,['Protein Abundance', 'mRNA Abundance']])
    sns.jointplot('Protein Abundance', 'mRNA Abundance', data=lgData, kind='kde', space=0)
```

```python
with sns.axes_style("white"):
    posts = yeastData['Transcription Rate'] > 0
    postl = yeastData['Translation Rate'] > 0
    lgData = np.log10(yeastData.loc[posts&postl,['Transcription Rate', 'Translation Rate']])
    sns.jointplot('Transcription Rate', 'Translation Rate', data=lgData, kind='kde', space=0)
```

```python
genepfxs = ['hxt', 'snf3']

mask = []
for n in yeastData['Gene Name']:
    if hasattr(n, "lower") and any(n.lower().startswith(x) for x in genepfxs):
        mask.append(True)

qgrid.show_grid(yeastData.loc[[ ('hxt' in n.lower()) if hasattr(n,"lower") else False for n in yeastData['Gene Name']],['Gene Name', 'Description', 'Protein Abundance', 'mRNA Abundance']])
```
